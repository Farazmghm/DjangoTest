```python
import pandas as pd

# ایجاد یک دیتاست فرضی
data = {
    'Age': [25, 30, 45, 50, 60, 35, 40, 55, 23, 38],
    'Driving_Experience': [7, 12, 25, 30, 40, 15, 20, 35, 5, 18],
    'Accidents_Count': [2, 0, 1, 0, 1, 1, 2, 0, 3, 1],
    'Previous_Insurance_Cost': [500, 450, 600, 700, 800, 550, 620, 750, 480, 590],
    'Insurance_Cost': [520, 460, 620, 710, 820, 560, 640, 770, 500, 610]
}

df = pd.DataFrame(data)
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np

# انتخاب متغیرهای مستقل
X = df[['Age', 'Driving_Experience', 'Accidents_Count', 'Previous_Insurance_Cost']]

# افزودن یک ستون ثابت برای محاسبه VIF
X['Intercept'] = 1

# محاسبه VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# حذف ستون ثابت از جدول
vif_data = vif_data[vif_data["Feature"] != "Intercept"]
print(vif_data)
```
